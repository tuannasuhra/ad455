```r

library(quantmod)
library(tidyverse)
####
getSymbols("ITXE.XC")
inditex<- as.data.frame(ITXE.XC)
inditex
```

```r
inditex %>% select(ITXE.XC.Open, ITXE.XC.Close) #Make a select operation of columns
```

```r
inditex %>% select(-ITXE.XC.High, -ITXE.XC.Low, -ITXE.XC.Volume, -ITXE.XC.Adjusted)
```

```r
#Make a filter operation of rows 
inditex %>%
filter(!is.na(ITXE.XC.Open))
```

```r
#Make a mutate operation to create a new column

inditex2<- inditex %>%
mutate(inditex_margin = (ITXE.XC.Close-ITXE.XC.Open) ) %>%
mutate(datess= rownames(inditex), Code=  1:nrow(inditex2))

inditex2
```

```r
#Make a group_by and summarize operation
inditex2 %>%
summarise(max_margin = max(inditex_margin, na.rm = T))
```

```r
inditex2 %>%
group_by(datess,Code) %>%
summarise(max_margin = max(inditex_margin, na.rm = T))
```

```r
 #Make a pivot_longer operation
#Make a pivot_wider operation
inditex_wider <- inditex2 %>%
group_by(datess, Code) %>%
summarise(max_marginn = max(inditex_margin, na.rm = T))
inditex_wider
```

```r
inditex_at_wide <- inditex_wider %>% pivot_wider(id_cols = Code , names_from = datess, values_from = max_marginn)
inditex_at_wide
```

```r
library(magrittr)
```

```r
inditex_wider %<>% 
mutate_at("Code", as.character)
```

```r
inditex_wider %>% 
pivot_longer(cols = -"max_marginn", names_to = "datess", values_to = "Code", values_drop_na = T)
```

```r
#Make a left_join operation with another table with common columns (you can create a new table out of the original one for this purpose, if you don't have a second table)
getSymbols("AAPL")
apple<- as.data.frame(AAPL)
apple %<>% 
mutate(Code2= 1:nrow(apple),datess2= rownames(apple),applemargin= AAPL.Open-AAPL.Close)
```
